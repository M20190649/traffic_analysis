### 處理無資料部分



```python
sec_info = pd.read_csv('SECINFO.csv')
#display(sec_info.head())
sec_info = sec_info.dropna()
sec_cols = forecast_df.columns
sectid_ids_gt = list(sec_info['SECTORID'])
not_in_gt = 0
in_gt = 0
sec_cols_gt = list()
for sec in sec_cols:
    if sec not in sectid_ids_gt:
        not_in_gt += 1
        #display('In sectodID: ' + sec)
    else:
        in_gt +=1
        sec_cols_gt.append(sec)
        #display(sec)
        
display('In forecast but not in ground_truth: ' + str(not_in_gt))
display('In forecast and ground truth: ' +str(in_gt))
```

```python
forecast_df_refine = forecast_df.set_index('time')
forecast_df_refine = forecast_df_refine[sec_cols_gt]
display(forecast_df_refine.head())
```

```python
profile = ProfileReport(forecast_df_refine, minimal=True, title='道路預測報告', html={'style':{'full_width':True}}).to_file(output_file="your_report.html")
```
